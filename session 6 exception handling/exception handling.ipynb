### TRY AND EXCEPT blocks

```python
def divide_handled():
     try:
          c = float(input('please enter the numerator'))
          d = float(input('please enter the denominator')) 
          div = c/d 
          return div
     except Exception as e:
         print(f'exception occures :{e} ')
```

```python
divide_handled()
```

```python
divide_handled()
print(9*4)
```

```python
def div_hand2():
    try:
        a = float(input('Please Enter Numerator : '))
        b = float(input('Please Enter Denominator : '))
        div = a/b
        return div
    except ZeroDivisionError:
        print('Cannot divide by 0, give any other number in denominator')
    except ValueError:
        print('Please do not enter alphabets instead of numbers')
```

```python
div_hand2()
```

```python
div_hand2()
```

### handling input exception 

```python
try:
    age = int(input('please enter your age : '))
    if age < 0 and age > 100:
        raise ValueError("age should be between 0 and 100")
    print(f'age of {age} year is a VALID AGE')
except:
    print(f'INVALID AGE {age} was entered')
```

### try except finally block

```python
from math import sqrt
```

```python
def sqr_root(n):
    try:
        from math import sqrt
        s = sqrt(n)
        print(f'Square root of given number {n} is {s:.4f}')
    except Exception as e:
        print(f'Exception Occured : {e}')
    finally:
        print('This block is always excuted Finally')
```

```python
sqr_root(25)
```

```python
def quadratic_roots(a,b,c):
    try:
        from math import sqrt
        d = b**2 - 4*a*c
        x1 = (-b+sqrt(d))/(2*a)
        x2 = (-b-sqrt(d))/(2*a)
        print(f'Real Roots of equation found : {x1:.4f} and {x2:.4f}')
        return x1, x2
    except Exception as e:
        print(f'Exception Occured : {e}')
    finally:
        print('Thank you for using this function')
```

```python
quadratic_roots(1,-5,6)
```
